# Read data and look up correlation matrix

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
import csv
```

```python
# Read data 
df = pd.read_csv('2020PM10.csv')
df = df.drop(columns=['FECHA','HORA'])
df.corr()
```

# Filter features that cannot be imputed

```python
# generate histograms of features removing invalid values
plt.rcParams.update({'figure.max_open_warning': 0})
for column in df:
    col_nan = df[column].replace(-99, np.nan)
    # Percentage of data per attribute
    percentage = (1 - col_nan.isna().sum()/(col_nan.count()+col_nan.isna().sum()))*100
    print(f'{column} = {percentage}%')
    col_clean = col_nan.dropna()
    if len(col_clean.index) != 0:
        fig, axi = plt.subplots()
        axi.set_title(str(column))
        col_clean.hist(bins=50, ax=axi)
```

```python
# look up table that considers attributes with data percentage availability > 50% or not too spread distribution
lookup_table = [
    'ACO',
    'AJM',
    'ATI',
    'BJU',
    'CAM',
    'CUA',
    'CUT',
    'FAC',
    'GAM',
    'ITZ',
    'MER',
    'PED',
    'SAG',
    'SFE',
    'TAH',
    'TLA',
    'UIZ',
    'VIF'
]
```

# Random Imputation using uniform distribution

```python
# generate random numbers from uniform distribution with min max values per attribute
unif_dist_df = pd.DataFrame(index=df.index)
final_df = pd.DataFrame(index=df.index)
for column in df:
    if column in lookup_table:
        col_nan = df[column].replace(-99, np.nan)
        final_df[column]=col_nan
        col_clean = col_nan.dropna()
        if len(col_clean.index) != 0:
            fig, axi = plt.subplots()
            axi.set_title(str(column))
            minimum = col_clean.min()
            maximum = col_clean.max()

            rnd = pd.Series(np.random.randint(minimum, maximum, col_nan.isna().sum()))
            col_nan[col_nan.isna()] = rnd.values
            col_nan.hist(bins=50, ax=axi)
            unif_dist_df[column]=col_nan
```

# Random Imputation using uniform distribution between first and third quartile

```python
# generate random numbers from uniform distribution with min max values per attribute
# now min max values will be narrowed down to the 25th and 75th percentile
unif_dist_perc_df = pd.DataFrame(index=df.index)
for column in df:
    if column in lookup_table:
        col_nan = df[column].replace(-99, np.nan)
        col_clean = col_nan.dropna()
        if len(col_clean.index) != 0:
            fig, axi = plt.subplots()
            axi.set_title(str(column))
            minimum = np.percentile(col_clean, 25)
            maximum = np.percentile(col_clean, 75)

            rnd = pd.Series(np.random.randint(minimum, maximum, col_nan.isna().sum()))
            col_nan[col_nan.isna()] = rnd.values
            col_nan.hist(bins=50, ax=axi)
            unif_dist_perc_df[column]=col_nan
```

# Random Imputation using log-normal distribution

```python
# generate random numbers from lognormal distribution
lognorm_dist_df = pd.DataFrame(index=df.index)
for column in df:
    if column in lookup_table:
        col_nan = df[column].replace(-99, np.nan)
        col_clean = col_nan.dropna()
        # ML estimator for lognormal dist
        mu = (col_clean.apply(lambda x: np.log(x)).sum())/col_clean.count()
        sigma = (col_clean.apply(lambda x: np.power(np.log(x)-mu, 2)).sum())/col_clean.count()
        print(f'mu = {mu}, sigma={sigma}')
        if len(col_clean.index) != 0:
            fig, axi = plt.subplots()
            axi.set_title(str(column))
            rnd = pd.Series(np.random.lognormal(mu, sigma, col_nan.isna().sum()))
            col_nan[col_nan.isna()] = rnd.values
            col_nan.hist(bins=50, ax=axi)
            lognorm_dist_df[column]=col_nan
```

# Generate box plot per attribute and metrics to compare distributions


```python
for column in final_df:
    if column in lookup_table:
        stats = pd.DataFrame(index=['count', 'mean', 'std', 'min', '25%', '75%', 'max'])
        
        stats['original_'+ column] = final_df[column].describe()
        stats['uniform_'+ column] = unif_dist_df[column].describe()
        stats['uniform_range_'+ column] = unif_dist_perc_df[column].describe()
        stats['lognorm_'+ column] = lognorm_dist_df[column].describe()
        
        col_final_df = final_df[column]
        col_unif_dist_df = unif_dist_df[column]
        col_unif_quart_df = unif_dist_perc_df[column]
        col_lognorm_df = lognorm_dist_df[column]
        
        fig, axs = plt.subplots(4, 2, figsize=(20, 15))
        axs[0,0].set_title('Original Distribution')
        axs[1,0].set_title('Random Uniform Distribution')
        axs[2,0].set_title('Random Uniform Distribution at 1st & 3rd quartile')
        axs[3,0].set_title('Random Lognorm Distribution')
        col_final_df.hist(bins=50, ax=axs[0,0])
        col_final_df.plot.box(ax=axs[0,1])
        
        col_unif_dist_df.hist(bins=50, ax=axs[1,0])
        col_unif_dist_df.plot.box(ax=axs[1,1])
        
        col_unif_quart_df.hist(bins=50, ax=axs[2,0])
        col_unif_quart_df.plot.box(ax=axs[2,1])
        
        col_lognorm_df.hist(bins=50, ax=axs[3,0])
        col_lognorm_df.plot.box(ax=axs[3,1])
        
        print(stats)
        
        
        
```

# Normalization

```python
for column in lognorm_dist_df:
    minmax_values = []
    min_val = lognorm_dist_df[column].min()
    max_val = lognorm_dist_df[column].max()
    mean = lognorm_dist_df[column].mean()
    std_dev = lognorm_dist_df[column].std()
    lognorm_minmax = lognorm_dist_df[column].apply(lambda x: (x-min_val)/(max_val-min_val))
    lognorm_z_score = lognorm_dist_df[column].apply(lambda x: (x-mean)/(std_dev))
    
    stats = pd.DataFrame(index=['count', 'mean', 'std', 'min', '25%', '75%', 'max'])

    stats['original_'+ column] = lognorm_dist_df[column].describe()
    stats['minmax_'+ column] = lognorm_minmax.describe()
    stats['zscore_'+ column] = lognorm_z_score.describe()

    col_final_df = lognorm_dist_df[column]

    fig, axs = plt.subplots(3, 2, figsize=(20, 15))
    axs[0,0].set_title('Original Distribution')
    axs[1,0].set_title('MinMax Distribution')
    axs[2,0].set_title('Z-Score  Distribution')
    
    col_final_df.hist(bins=50, ax=axs[0,0])
    col_final_df.plot.box(ax=axs[0,1])

    lognorm_minmax.hist(bins=50, ax=axs[1,0])
    lognorm_minmax.plot.box(ax=axs[1,1])

    lognorm_z_score.hist(bins=50, ax=axs[2,0])
    lognorm_z_score.plot.box(ax=axs[2,1])

    print(stats)
        
    
    
```
